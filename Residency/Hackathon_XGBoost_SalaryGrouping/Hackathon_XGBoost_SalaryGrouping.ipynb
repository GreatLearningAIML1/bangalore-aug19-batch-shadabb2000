```python
df_drop = df.drop(['ID','Year Type','Employee Identifier','Organization Group','Department','Union','Job Family','Job'], axis=1)
test_df_drop = test_df.drop(['ID','Year Type','Employee Identifier','Organization Group','Department','Union','Job Family','Job'], axis=1)
df_drop.head()
```

```python
## Label Encoding of all the columns
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()

# Categorical boolean mask
categorical_feature_mask = df_drop.dtypes==object
# filter categorical columns using mask and turn it into a list
categorical_cols = df_drop.columns[categorical_feature_mask].tolist()
df_drop[categorical_cols] = df_drop[categorical_cols].apply(lambda col: le.fit_transform(col))
print(df_drop.info())

# Categorical boolean mask
categorical_feature_mask = test_df_drop.dtypes==object
# filter categorical columns using mask and turn it into a list
categorical_cols = test_df_drop.columns[categorical_feature_mask].tolist()
test_df_drop[categorical_cols] = test_df_drop[categorical_cols].apply(lambda col: le.fit_transform(col))
print(test_df_drop.info())

```

```python
from xgboost import XGBClassifier
# from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
# shape of the dataset
print('Shape of training data :',df_drop.shape)
print('Shape of testing data :',test_df_drop.shape)
```

```python
# seperate the independent and target variable on training data
train_X = df_drop.drop(columns=['Class'],axis=1)
train_y = df_drop['Class']
```

```python
test_X = test_df_drop
```

```python
model = XGBClassifier(n_estimators=1000, max_depth=8)

# fit the model with the training data
model.fit(train_X,train_y)
```

```python
# predict the target on the train dataset
predict_train = model.predict(train_X)
print('\nTarget on train data',predict_train) 
```

```python
# Accuray Score on train dataset
accuracy_train = accuracy_score(train_y,predict_train)
print('\naccuracy_score on train dataset : ', accuracy_train)
```

```python
# predict the target on the test dataset
predict_test = model.predict(test_X)
print('\nTarget on test data',predict_test) 
```
